```python
# Task 5: Data Analysis---EDA

# Univariate analysis : Single variable Analysis
# Analysis each of single variable and their data  distribution --- data distribution of each of the single columns 
```

```python
# Price Analysis
# price distribution of the 
AirBnB_data['price']
```

```python
# identifying the outliers --- box plot 
sns.boxplot(data= AirBnB_data, x = 'price' )

# this shows that most of the values comes under 20,000 and one outlier is comes under the range of 100,000
so we have to remove that 
```

```python
# removing the outliers-- create a new dataframe where price values less than 100000 and under 1500
new_AirBnB_data = AirBnB_data [AirBnB_data['price'] < 1500]
```

```python
# Now again creating box plot 
sns.boxplot(data= new_AirBnB_data, x = 'price' )
```

```python
# creating histogram with new data frame 

plt.figure(figsize = (8,5))

sns.histplot (data = new_AirBnB_data, x = 'price', bins = 100 )

plt.title('Price Distribution')
plt.ylabel ('Frequency')
plt.show()
```

```python
new_AirBnB_data.dtypes # this helps in check the colums and their data types
```

```python

plt.figure(figsize = (10,3))  # is used to set the dimensions of the figure in Matplotlib. 10 inches wide and 3 inches tall.

sns.histplot (data = new_AirBnB_data, x = 'availability_365' )

plt.title('availability_365_Distribution')
plt.ylabel ('Frequency')
plt.show()
```

```python
# Check the Avg price in each 'neighbourhood' 

# for this we will use groupby function ------> df.groupby('column_name').aggregation_function()

new_AirBnB_data.groupby( by = 'neighbourhood_group')['price'].mean()
```

```python
# Feature Engineering : 
# Feature engineering is the process of creating, modifying, or selecting features from raw data to improve the performance of machine learning models. 
# It plays a crucial role in data preprocessing, helping algorithms make better predictions.
```

```python
# creating a new column ----> Price per bed 
new_AirBnB_data.insert(21,'price_per_bed', (new_AirBnB_data['price'] / new_AirBnB_data['beds']))

```

```python
new_AirBnB_data.head()
```

```python
# calculate the neighbourhood group with average price per bed 
new_AirBnB_data.groupby (by = 'neighbourhood_group')['price_per_bed'].mean()
```

```python
# price depending on neighbourhood

sns.barplot(data = new_AirBnB_data , x = 'neighbourhood_group', y = 'price', hue = 'room_type')

# the hue parameter is used to add an additional categorical variable to your visualization.
# Specifically, it separates the bars based on the values of the hue variable and uses different colors to represent each category.
```

```python
# Number of reviews and price relationship
plt.figure(figsize = (10,5))
plt.title('Locality and Review Dependency')
sns.scatterplot ( data =  new_AirBnB_data , x = 'number_of_reviews', y = 'price', hue = 'neighbourhood_group')
```

```python
sns.pairplot(data = new_AirBnB_data, vars = ['price','minimum_nights','number_of_reviews', 'availability_365'], hue = 'room_type') 

# The sns.pairplot() function in Seaborn is used to create pairwise scatter plots for numerical variables in a dataset. 
# It’s a great way to visualize relationships between multiple numerical features at once.
```
